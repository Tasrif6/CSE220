```python
#Singly
#Cycle linkedList or not
def circular(a):
    head=None
    head=Node(a[0],None)
    temp=head.next
    tail=head
    while tail.next!=None:
        tail=tail.next
    tail.next=temp
    if tail.next==temp:
        print('Circular')
    else:
        print('Not Circular')

a=[2,3,4,1]
circular(a)
```
