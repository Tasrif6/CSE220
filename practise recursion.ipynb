```python
#Number-1:
def recar(n):
    # Base case: If the number has only one digit, return that digit
    if n < 10:
        return n
    
    # Recursive case: Sum the rightmost digit (n % 10) with the sum of the rest of the digits (n // 10)
    else:
        return n % 10 + recar(n // 10)
number=126
recar(number)
```

```python
def sumDigits(n):
    if n<10:
        return n
    else:
        result=n%10
        return result+sumDigits(n//10)
number=int(input())
sumDigits(number)
```

```python
def bunnyEars(n):
    if n==0:
        return n
    elif n%2==0:
        return 3+bunnyEars(n-1)
    elif n%2!=0:
        return 2+ bunnyEars(n-1)
number=int(input())
bunnyEars(number)
```
